# Chap 1 The Machine Learning Landscape

#### Machine learning:
Machine Learning is the field of study that gives computers the ability to learn
without being explicitly programmed.<br>
A computer program is said to learn from experience E with respect to some task T
and some performance measure P, if its performance on T, as measured by P, improves
with experience E.<br>

#### uses:
1. Spam filterss
2. Natural Language processing 

1. Whether or not they are trained with human supervision (supervised, unsuper‐vised, semisupervised, and Reinforcement Learning)

2. Whether or not they can learn incrementally on the fly (online versus batchlearning)

3. Whether they work by simply comparing new data points to known data points,or instead detect patterns in the training data and build a predictive model, much like scientists do (instance-based versus model-based learning)

#### Supervised learning:
In supervised learning, the training data you feed to the algorithm includes the `desired solutions, called labels`.

example: `Classification`: spam filter, `Regression` to predict a numerical value such as the price of a car,
given a set of features (mileage, age, brand, etc.) called predictors<br>
<b>In Machine Learning an attribute is a data type (e.g., “Mileage”),
while a feature has several meanings depending on the context, but
generally means an attribute plus its value</b><p>
`Logistic Regression` is a regression algorithm that is used for classification, it can
output a value that `corresponds to the probability of belonging to a given class` (e.g.,
20% chance of being spam). <br>
<br> 


Some important Supervised Learning Algorithms are:
1. k-Nearest Neighbors
2. Linear Regression
3. Logistic Regression
4. Support Vector Machines (SVMs)
5. Decision Trees and Random Forests
6. Neural networks 2

#### Unsupervised Learning:
`the training data is unlabeled, The system tries to learn without a teacher.`


`Example: Data analysis of Blog visitors:`
<b>Clustering algorithm</b> detects the groups<br>
`Hierarchical Cluster Analysis` may even sub divide the clusters into smaller clusters.<br>
`Dimensionality reduction`, in which the goal is to simplify(merge the several corelated features into one) the data without losing too much information.<br><p>

What this achieves is `Feature extraction`. This causes `less disk and memory usuage` and also better `performance`.<br><p>
    
another important unsupervised task is `Anomaly detection`. E.g. unusual credit card transactions to prevent fraud, catching manufacturing defects,
or automatically removing outliers from a dataset before feeding it to another learn‐
ing algorithm. `Trained with normal ones and thus it can detect an anomaly for new instances.`<br><p>

`Association rule learning` goal is to dig into large amounts of data and discover interesting relations between attributes. E.g. Supermarket logs---> People who buy barbecue sauce and potato chips also tend to buy steak, so put them close.

#### Semisupervised learning:
The algorithms can deal with `partially labeled` training data, usually a lot of unla‐
beled data and a little bit of labeled data.<br><p>
    
E.g. Google photos---> detects people in different photos(`Clustering(Unsupervised part)`) it needs you to label people for once and it can detect them in every other photos(`Supervised part`):


Examples: 
`Deep belief networks (DBNs)` are based on `unsu‐
pervised components` called `restricted Boltzmann machines (RBMs)` stacked on top of
one another. RBMs are trained sequentially in an unsupervised manner, and then the
whole system is fine-tuned using supervised learning techniques.

#### Reinforcement Learning:
There is an `agent`,which is `learning system` ---> performs actions---> gets `rewards` and `penalties` for +ve and -ve actions and then comes up with a `policy`, which is the best strategy(what this agent should do in particular situation to get most rewards).


Examples: Robots learning to walk, Deepmind's AlphaGo

#### Batch Learning:
system has to be `trained with available data`--> offline,more time, more resources needed.<br>
For `new set of data` it has to be trained from level 0 again for new version. so every time the change has to be made, it has to be automated to change accordingly and train itself.<br>
<p>
    
#### Online Learning:
train the system incrementally by feeding it data instances
sequentially, either individually or by small groups called mini-batches. on the `fly`<br><p>
    
Online learning is great for systems that receive data as a continuous flow (e.g., stock prices) and need to adapt to change rapidly or autonomously. good if resources are limited. And also to train systems on huge datasets that
    cannot fit in one machine’s main memory (this is called `out-of-core learning`). All of this is done `offline` so better name would be `incremental learning`.<p>
    
`learning rate` is a parameter of online learning systems is how fast they should adapt to changing data. High---> learns quickly but forgets the older ones also quickly, Low---> learns slowly, less sensitive to noise in the new data or to sequences of nonrepresentative data points.

<p>
    
In online learning if bad data is fed, performance declines. So the data and its sources has to be continuously monitored.

#### Instance-Based Versus Model-Based Learning:
ML can be categorized by how they `generalize`. 2 categories on basis of approach to generalization(system needs to be able to generalize to examples it has never seen before).<p>
    
##### Instance Based Learning:
The system learns the examples by heart, then generalizes to new cases using a similarity measure.

##### Model-Based Learning:
Build a model from examples, then use that model to make predictions.
Example:<br>
linear model with parameters A and B:<br>
`life_satisfaction = A + B * GDP_PER_CAPITA`, tuning A and B we get a model. To tune here we need to specify `Performace measure` either `Utility function(fitness function)---> Measure of how good your model is` or `Cost function---> Measure of how bad your function is` say for linear regression problems like one mentioned above, we use cost function and minimize the cost using the training examples. This is called `Training the model`. Training gives us values of A and B. Now we run the model for the predictions.
