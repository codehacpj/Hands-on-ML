`F1score`, in particular if you need a simple way to compare two classifiers. The F 1 score is
the harmonic mean of precision and recall.<p>
    
    
you can‚Äôt have it both ways: increasing precision reduces recall, and
vice versa. This is called the precision/recall tradeoff. And precision and recall have different importance according to what we want to achieve.<br>


<b>A high-precision classifier is
not very useful if its recall is too low!</b>

### The ROC Curve:
receiver operating characteristic (ROC) curve is another common tool used with
binary classifiers. It is very similar to the precision/recall curve, but instead of plot‚Äê
ting precision versus recall, the ROC curve plots the true positive rate (another name
for recall) against the false positive rate.
<p>
    
### Multiclass classifiers:
multiclass classifiers (also
called multinomial classifiers) can distinguish between more than two classes.<br>
e.g. Random Forest classifiers or naive Bayes classifiers<br>
`one-versus-all (OvA) strategy`<br>
`one-versus-one (OvO) strategy.`<br>
For Support Vector Machine classifiers OvO is preferred, for most OvA is preferred.<br><p>
